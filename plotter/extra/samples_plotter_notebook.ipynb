```python
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from mpl_toolkits.basemap import Basemap
import numpy as np
import pickle
import re
from os import walk
```

```python
def unpickle(path):
    with open(path, 'rb') as f:
        data = pickle.load(f, encoding='latin1')
    return data

def tryint(s):
    try:
        return int(s)
    except:
        return s
    
def alphanum_key(s):
    """ Turn a string into a list of string and number chunks.
        "z23a" -> ["z", 23, "a"]
    """
    return [ tryint(c) for c in re.split('([0-9]+)', s) ]

def sort_nicely(l):
    """ Sort the given list in the way that humans expect.
    """
    l.sort(key=alphanum_key)
    print('Sorted humanly')
    
def get_all_files(source_path, filename):
    fullpath = source_path + filename
    f = []
    for (dirpath, dirnames, filenames) in walk(fullpath):
        f.extend(filenames)
        break
    return f
```

```python
lat_path = './lats_32x32.pkl'
lon_path = './lons_32x32.pkl'

# lat,lon = grb.latlons() # Set the names of the latitude and longitude variables in your input GRIB file
lon = unpickle(lon_path)
lat = unpickle(lat_path)
print(len(lat))
print(len(lon))
```

```python
# should be in main? nah.
source_path = "./experiments_samples/32x32_temp_v1/"
filename = 'samples/'
endfile = 'plots/'

file_names = get_all_files(source_path=source_path, filename=filename)
sort_nicely(file_names)
paths = []

futu = []
orig = []
for file in file_names:
    if 'future' in file:
        futu.append(source_path + filename + file)
    elif 'image' in file:
        orig.append(source_path + filename + file)
        
print(orig)
```

```python
counter = 0
batch_size = 64
for index in range(batch_size):
    for future, original in zip(futu,orig):
        path = original
        path2 = future
        
        print(path, path2)
        # LOAD ORIGINAL
        data = unpickle(path)
        print(data.shape)

        data = data[index]
        data = data.reshape((32,32))
        print(data.shape)
        print(max(data[0]))
        print(min(data[0]))

        # LOAD PREDICTION
        data2 = unpickle(path2)
        print(data2.shape)

        data2 = data2[index]
        data2 = data2.reshape((32,32))
        print(data2.shape)
        print(max(data2[0]))
        print(min(data2[0]))
        
        fig = plt.figure(dpi=500)

        ax = fig.add_subplot(211)
        ax.set_title("Original")
        m = Basemap(projection='mill', llcrnrlon=13,urcrnrlon=22,llcrnrlat=45,urcrnrlat=54.3,resolution='h')
        x, y = m(lon,lat)
        m.pcolormesh(x,y,data)
        m.drawmapboundary()
        m.drawcoastlines()
        m.drawcountries()

        ax = fig.add_subplot(221)
        ax.set_title("Forecast")
        m = Basemap(projection='mill', llcrnrlon=13,urcrnrlon=22,llcrnrlat=45,urcrnrlat=54.3,resolution='h')
        x, y = m(lon,lat)
        m.pcolormesh(x,y,data2-273.15)
        m.drawmapboundary()
        m.drawcoastlines()
        m.drawcountries()
        
        # fig.colorbar(pad=0.10, label='Teplota')
        # fig.title('Teplota') # Set the name of the variable to plot
        # 
        plt.savefig(source_path + endfile + 'temp_32_batch' + str(index) + '_pic' + str(counter) + '.png', bbox_inches = "tight") # Set the output file name
        counter = counter + 1
        plt.show()
```
